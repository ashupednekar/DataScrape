```python
import json
with open('../acerph/p1.json') as f:
    data = json.loads(f.read())
```

```python
for k, v in data.items():
    if k=='Laptop':
        category = 'notebooks'
    else:
        category = k.lower()
    for a, b in v['sub-categories'].items():
        if 'items' in b:
            for x in b['items']:
                sub_product = x['name'].lower()
                if sub_product in ['aopen 24ml1y', 'aopen 24hc1qr pbidpx']:
                    sub_category = 'affordable'
                elif sub_product in ['helios 300 ph315-53-73rt', 'nitro 5 an515-43-r2wk', 'nitro 7 an715-51-58x1', 'predator triton 500 pt515-52-75zy']:
                    sub_category = 'high performance'
                elif sub_product in ['aspire 3 a315-56-596k', 'aspire 5 a514-52k-39ad']:
                    sub_category = 'best sellers'
                elif sub_product in ['acer x1126ah', 'c250i']:
                    sub_category = 'wider displays'
                else:
                    sub_category = a.lower()
                if sub_product in ['acer x1126ah', 'c250i']:
                    product = 'wider displays'
                elif len(sub_product.split(' ')) > 1:
                    if len(sub_product.split(' ')[1]) == 1:
                        #aspire 5 jbvdkjbjkv
                        product = ' '.join(sub_product.split(' ')[:2])
                    elif '-' in sub_product.split(' ')[1]:
                        #aspire tc-djdkslvdlk
                        product = sub_product.split('-')[0]
                else:
                    product = sub_product
                buy_now_url = x['link']
                image_url = x['image'] #TODO: call picsum api
                print(category + '|' + product + '|' + sub_category + '|' + sub_product + '|' + buy_now_url + '|' + image_url)
```
